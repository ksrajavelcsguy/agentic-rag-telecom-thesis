### Loading config + setting safe defaults

```python
import os, json, time, random
import numpy as np

BASE_DIR = "/content/drive/MyDrive/agentic-rag-telecom-thesis"
DOC2DIAL_DIR = os.path.join(BASE_DIR, "data/raw/doc2dial")
RESULTS_DIR = os.path.join(BASE_DIR, "results")
TRACE_DIR = os.path.join(RESULTS_DIR, "traces")
os.makedirs(RESULTS_DIR, exist_ok=True)
os.makedirs(TRACE_DIR, exist_ok=True)

DOC_FILE  = os.path.join(DOC2DIAL_DIR, "doc2dial_doc.json")
DIAL_FILE = os.path.join(DOC2DIAL_DIR, "doc2dial_dial_validation.json")

# knobs
seed = 42
n_pairs = 50
chunk_size = 1000
chunk_overlap = 150
top_k = 5

print("DOC_FILE exists:", os.path.exists(DOC_FILE))
print("DIAL_FILE exists:", os.path.exists(DIAL_FILE))
```

### Loading ALL documents correctly (domain → doc_id → doc)

```python
docs_json = json.load(open(DOC_FILE, "r", encoding="utf-8"))

def iter_docs(obj):
    # doc2dial_doc.json is indexed by domain -> doc_id -> doc_object
    if isinstance(obj, dict):
        # if this dict looks like a doc
        if "doc_id" in obj and ("doc_text" in obj or "spans" in obj):
            yield obj
            return
        for v in obj.values():
            yield from iter_docs(v)
    elif isinstance(obj, list):
        for x in obj:
            yield from iter_docs(x)

doc_items = []
for d in iter_docs(docs_json):
    doc_id = d.get("doc_id")
    txt = d.get("doc_text") or ""
    if doc_id and isinstance(txt, str) and len(txt.strip()) > 200:
        doc_items.append({"doc_id": str(doc_id), "text": txt})

print("✅ Docs loaded:", len(doc_items))
print("Sample doc_id:", doc_items[0]["doc_id"], "| chars:", len(doc_items[0]["text"]))
```

```python
print(type(docs_json), list(docs_json.keys())[:5] if isinstance(docs_json, dict) else "not dict")
```

### Chunking + indexing into Chroma (persisting in Drive)

```python
import chromadb
from sentence_transformers import SentenceTransformer

def chunk_text(text, size=1000, overlap=150):
    chunks = []
    start = 0
    while start < len(text):
        end = min(len(text), start + size)
        chunks.append(text[start:end])
        start = end - overlap
        if start < 0: start = 0
        if end == len(text): break
    return chunks

chunks, metadatas, ids = [], [], []
for d in doc_items:
    for j, ch in enumerate(chunk_text(d["text"], chunk_size, chunk_overlap)):
        ids.append(f'{d["doc_id"]}__{j}')
        chunks.append(ch)
        metadatas.append({"doc_id": d["doc_id"], "chunk_id": j})

CHROMA_DIR = os.path.join(RESULTS_DIR, "chroma_baseline")
client = chromadb.PersistentClient(path=CHROMA_DIR)

try:
    client.delete_collection("doc2dial_baseline")
except Exception:
    pass

col = client.create_collection("doc2dial_baseline")
embedder = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

BATCH = 256
for i in range(0, len(chunks), BATCH):
    batch_text = chunks[i:i+BATCH]
    embs = embedder.encode(batch_text, normalize_embeddings=True).tolist()
    col.add(
        ids=ids[i:i+BATCH],
        documents=batch_text,
        metadatas=metadatas[i:i+BATCH],
        embeddings=embs
    )

print("✅ Chunks indexed:", len(chunks), "| CHROMA_DIR:", CHROMA_DIR)
```

### Extracting user→agent pairs correctly (fixing “0 pairs”)

```python
dial_json = json.load(open(DIAL_FILE, "r", encoding="utf-8"))

def iter_dialogues(obj):
    if isinstance(obj, dict):
        if "turns" in obj and isinstance(obj["turns"], list):
            yield obj
            return
        for v in obj.values():
            yield from iter_dialogues(v)
    elif isinstance(obj, list):
        for x in obj:
            yield from iter_dialogues(x)

def extract_user_agent_pairs(dial_obj):
    pairs = []
    for d in iter_dialogues(dial_obj):
        turns = d.get("turns", [])
        last_user = None
        for t in turns:
            role = (t.get("role") or "").lower()
            utt = t.get("utterance")
            if not isinstance(utt, str):
                continue
            utt = utt.strip()
            if not utt:
                continue
            if role == "user":
                last_user = utt
            elif role == "agent" and last_user:
                pairs.append((last_user, utt))
                last_user = None
    return pairs

pairs = extract_user_agent_pairs(dial_json)
random.seed(seed)
random.shuffle(pairs)
pairs = pairs[:n_pairs]

print("QA pairs for eval:", len(pairs))
print("Sample pair:", pairs[0] if pairs else "NONE")
```

### Generator WITHOUT pipeline task error (FLAN-T5 direct)

```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

device = "cuda" if torch.cuda.is_available() else "cpu"
tok = AutoTokenizer.from_pretrained("google/flan-t5-base")
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base").to(device)

def generate_text(prompt, max_new_tokens=256):
    inputs = tok(prompt, return_tensors="pt", truncation=True, max_length=2048).to(device)
    with torch.no_grad():
        out = model.generate(**inputs, max_new_tokens=max_new_tokens)
    return tok.decode(out[0], skip_special_tokens=True).strip()

print("Generator ready on:", device)
```

### RAG answer + metrics + saving results

```python
import evaluate
rouge = evaluate.load("rouge")
bleu  = evaluate.load("sacrebleu")

def retrieve(query, k=5):
    q_emb = embedder.encode([query], normalize_embeddings=True).tolist()[0]
    res = col.query(query_embeddings=[q_emb], n_results=k, include=["documents","metadatas"])
    return res["documents"][0], res["metadatas"][0]

def rag_answer(query, k=5):
    t0 = time.time()
    ctx_docs, metas = retrieve(query, k=k)
    context = "\n\n".join([f"[{m['doc_id']}#{m['chunk_id']}] {d}" for d, m in zip(ctx_docs, metas)])
    prompt = (
        "Answer the QUESTION using only the CONTEXT. "
        "If the answer is not in the context, say: I don't know.\n\n"
        f"CONTEXT:\n{context}\n\nQUESTION: {query}\nANSWER:"
    )
    out = generate_text(prompt, max_new_tokens=256)
    dt = time.time() - t0
    return out, dt, metas

preds, refs, lats = [], [], []
trace_path = os.path.join(TRACE_DIR, "baseline_traces.jsonl")

with open(trace_path, "w", encoding="utf-8") as f:
    for i, (q, ref) in enumerate(pairs, 1):
        ans, dt, metas = rag_answer(q, k=top_k)
        preds.append(ans)
        refs.append(ref)
        lats.append(dt)
        f.write(json.dumps({
            "i": i, "query": q, "prediction": ans, "reference": ref,
            "latency_s": dt, "top_k": top_k,
            "retrieved": [{"doc_id": m["doc_id"], "chunk_id": m["chunk_id"]} for m in metas]
        }, ensure_ascii=False) + "\n")

rou = rouge.compute(predictions=preds, references=refs, use_stemmer=True)
ble = bleu.compute(predictions=preds, references=[[r] for r in refs])

lat = np.array(lats)
lat_mean = float(lat.mean())
lat_med  = float(np.median(lat))
lat_p95  = float(np.percentile(lat, 95))

print("ROUGE-L:", rou["rougeL"])
print("BLEU:", ble["score"])
print("Latency mean/median/p95:", lat_mean, lat_med, lat_p95)
print("Traces saved:", trace_path)

# Save summary CSV for Results Ledger
import csv
out_csv = os.path.join(RESULTS_DIR, "baseline_summary.csv")
with open(out_csv, "w", newline="", encoding="utf-8") as fp:
    w = csv.DictWriter(fp, fieldnames=["system","rougeL","bleu","lat_mean_s","lat_median_s","lat_p95_s","n_pairs","top_k","chunk_size","chunk_overlap"])
    w.writeheader()
    w.writerow({
        "system":"Baseline RAG (Doc2Dial)",
        "rougeL":rou["rougeL"],
        "bleu":ble["score"],
        "lat_mean_s":lat_mean,
        "lat_median_s":lat_med,
        "lat_p95_s":lat_p95,
        "n_pairs":len(pairs),
        "top_k":top_k,
        "chunk_size":chunk_size,
        "chunk_overlap":chunk_overlap
    })
print("Saved:", out_csv)
```
